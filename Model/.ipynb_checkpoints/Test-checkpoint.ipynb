```python
from rdkit import Chem
from rdkit.Chem import PandasTools
from tqdm import tqdm
import openbabel

def iupac_to_smiles(names_list):
    obConversion = openbabel.OBConversion()
    obConversion.SetInAndOutFormats("iupac", "smi")
    mol = openbabel.OBMol()

    smiles_list = []
    for name in tqdm(names_list, desc="Converting IUPAC to SMILES", unit="name"):
        try:
            obConversion.ReadString(mol, name)
            smiles_string = obConversion.WriteString(mol).strip()
            if smiles_string:
                smiles_list.append(smiles_string)
            else:
                smiles_list.append(None)
                print(f"IUPAC name {name} could not be converted.")
        except Exception as e:
            print(f"Error for IUPAC name {name}: {e}")
            smiles_list.append(None)
    return smiles_list

# 示例
names_list = GC['Name'].values.tolist()
resulting_smiles = iupac_to_smiles(names_list)
print(resulting_smiles)
```

```python
import pubchempy as pcp

def get_smiles_from_cas(cas_number):
    try:
        compound = pcp.get_compounds(cas_number, 'name', record_type='compound', namespace='cas')[0]
        return compound.canonical_smiles
    except Exception as e:
        print(f"Error obtaining SMILES for CAS {cas_number}: {e}")
        return None

def get_smiles_from_iupac(iupac_name):
    try:
        compound = pcp.get_compounds(iupac_name, 'name', record_type='compound')[0]
        return compound.canonical_smiles
    except Exception as e:
        print(f"Error obtaining SMILES for IUPAC name {iupac_name}: {e}")
        return None

# 示例
cas_smiles = get_smiles_from_cas('50-78-2')  # CAS for Aspirin
iupac_smiles = get_smiles_from_iupac('Acetylsalicylic acid')  # IUPAC name for Aspirin

print(f"SMILES for CAS 50-78-2: {cas_smiles}")
print(f"SMILES for IUPAC Acetylsalicylic acid: {iupac_smiles}")
```
